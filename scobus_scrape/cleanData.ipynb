```python
df['Author Keywords'] = df['Author Keywords'].str.replace(';', ',', regex=False)
```

```python
import pycountry
import pandas as pd

# ฟังก์ชันเพื่อดึงชื่อประเทศจากข้อความ
def extract_countries_pycountry(affiliations):
    if pd.isna(affiliations):  # ตรวจสอบว่าเป็น NaN หรือไม่
        return None
    
    # สร้างลิสต์ของประเทศทั้งหมดจาก pycountry
    countries = [country.name for country in pycountry.countries]
    
    # ใช้ Regular Expression เพื่อตรวจหาชื่อประเทศในข้อความ
    found_countries = [country for country in countries if country in affiliations]
    
    return ', '.join(found_countries) if found_countries else None

# ใช้ฟังก์ชัน extract_countries_pycountry กับคอลัมน์ 'Authors with affiliations'
df['Authors Country'] = df['Authors with affiliations'].apply(extract_countries_pycountry)
```

```python
df[['Title','Source title','Cited by','Authors Country','Author Keywords', 'Year']].to_csv('100_000_filtered.csv', index=False)
```
